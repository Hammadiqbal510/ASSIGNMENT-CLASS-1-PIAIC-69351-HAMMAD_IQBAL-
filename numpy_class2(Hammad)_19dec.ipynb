# Combines two arrays in a manner where the columns of the second array are placed to the right of those in the first array. The function actually stacks the two items provided in a two-element tuple. The result is a new array with data copied from the two that are specified:

```python
import numpy as np
t=np.arange(16).reshape(4,4)
t
```

```python
w=np.arange(16).reshape(4,4)
w
```

```python
# note : horizental stack ma number of column increase hota ( h,c )
np.hstack((t,w)) # combine array and write it in horizental way
```

```python
# same process we can implement by using "concatenate" function
np.concatenate((t,w),axis=1)
# but in this case we can tell it ka hum apna array kis axis pa chaya
# horizental chahya to axis=1 kara ga bcz "x" axis pa tamam value horizental hoti
#vartical  chahya to axis=0 kara ga bcz "y" axis pa tamam value vertiacal hoti
```

```python
# note : vartical stack ma number of rows increase hota ( v,r )
np.vstack((t,w)) # combine array and write it in vartical way
```

```python
# same process we can implement by using "concatenate" function
np.concatenate((t,w),axis=0)
# but in this case we can tell it ka hum apna array kis axis pa chaya
# horizental chahya to axis=1 kara ga bcz "x" axis pa tamam value horizental hoti
#vartical  chahya to axis=0 kara ga bcz "y" axis pa tamam value vertiacal hoti
```

```python
# note : row   stack ma number of rows increase hota ( r,r )
np.row_stack((t,w)) # combine array and write it in vertial way
# row stack= vertical stack
```

```python
# note : column stack ma number of column increase hota ( v,r )
np.column_stack((t,w)) # combine array and write it in horizantal way
# row stack= horizatntal stack
```

```python
np.dstack((t,w))# note : depth stack ma number of deptth increase hota ( v,r )
# combine in depth array ka upr  array aya gi ek list pa dosri list fit us pa next list
```

# Manipulating and Saving the Image
Now that we have converted our image into a Numpy array, we might come across a case where we need to do some manipulations on an image before using it into the desired model. In this section, you will be able to build a grayscale converter. You can also resize the array of the pixel image and trim it.

After performing the manipulations, it is important to save the image before performing further steps. The format argument saves the file in different formats, such as PNG, GIF, or PEG.

For example, the code below loads the photograph in JPEG format and saves it in PNG format.

```python
import numpy as np
from PIL import Image

myImg = Image.open('hammad.jpg')
myImg = myImg.convert('L') #"L " convert into gray color
# explore it convert()shift +tab u can see more paramater
myImg = np.array(myImg)
# ya kam ek line ma b ho sktaa haaaa
#myImg = np.array(Image.open('hammad.jpg').convert())

print(type(myImg))

gray_img= Image.fromarray(myImg).save('gray_hammad.jpg')
gray_img
# balck and white image da da ga 
```
